### 1-Handling Null values & Duplicates

```python
df.isnull().sum()
```

#### 1.2- Removing Null Values

```python
df = df.dropna(subset =['job_id', 'job_role', 'company', 'experience', 'salary', 'location',
         'resposibilities', 'posted_on', 'job_link',
       'company_link'])
df
```

#### 1.3- Checking duplicates

```python
df.duplicated('job_link').sum()
```

```python
df.duplicated('job_id').sum()
```

####1.4- Removing Duplicates based on job_link column

```python
df =df.drop_duplicates(subset = 'job_link')
df
```

####3.1- Top 20 frequent responsibilities

```python
df_res =df['resposibilities'].str.lower().str.split(',').explode().value_counts().sort_values(ascending = False).reset_index().head(20)

sns.barplot(x ='resposibilities', y='count', data =df_res)
plt.title('top 20 skills')
plt.xlabel('Responsibilities')
plt.ylabel('frequency')
plt.xticks(rotation =90)
plt.show()
```

####3.2- Top 5 frequent responsibilities for any role

```python
resp_df= pd.DataFrame([['sales' ,	5722],
	['communication skills',	5585],
	['team handling',	2596],
	['management',	2099],
	['team management',	1942]])
resp_df.columns =['resposibilities', 'count']


sns.barplot(x ='resposibilities' , y='count', data= resp_df)
plt.title('top 5 skills')
plt.xlabel('Responsibilities')
plt.ylabel('count')
plt.xticks(rotation =90)
plt.show()
```

#### 3.2- Top 20 frequent responsibilities for specific company
