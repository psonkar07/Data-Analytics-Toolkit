### 2- Data Cleaning

####2.1- Remove the duplicates values as per language.

```python
len(df[df['language'] != 'Old']['link'])
```

```python
df.reset_index()
```

####2.2- Convert duration column into seconds.

```python
for i in df['duration'][:10]:
  lst = i.split(':')
  print(i,int(lst[0])*60 + int(lst[1]) )
```

#### Add into dataset.

```python
duration =[]
for i in df['duration']:
  lst = i.split(':')
  if (len(lst))== 2:
   duration.append(int(lst[0])*60 +int(lst[1]))
  else:
    duration.append(int(lst[0])*60*60 +int(lst[1])*60 +int(lst[2]))

#print(durations)
df['duration']= duration
```

####2.3- Remove '/artist/' into singer_id column.

```python
df['singer_id'] = df['singer_id'].str.replace('/artist/','').str.replace('|',',')
df.head()
```

####2.4-  Remove '|' operator in singer column.

```python
df['singer'] = df['singer'].str.replace('|', ',')
df.head()
```

####2.5-  Remove '/' operator in link column.

```python
df['link'] = df['link'].str.replace('/','')
df.head()
```

#### 3.1- find the number of unique singers.

```python
lst =[]
for i in df['singer_id']:
  lst.extend(i.split(','))
print(len(set(lst)))

df['singer_id'].str.split(',').explode().nunique()

len(list(set(','.join(df['singer_id']).split(','))))
```

```python
#lst =[]
for m in range(len(lst))[:10]:
 c= 0
 for i in df['singer_id']:
  if (lst[m] in i):
    c+=1
 print(lst[m],c)
```

####3.2- Top 10 singers who have sang max no. of songs.

```python
from tqdm import tqdm

singers = list(set(lst))

singer_lst = []

for singer in tqdm(singers):            # Going through every singer
  c = 0
  for i in df['singer_id']:       # Counting number of songs for each singer
    if (singer in i.split(',')):
      c += 1

  singer_lst.append([singer,c])

singer_df = pd.DataFrame(singer_lst, columns = ['singer','count'])

singer_df.sort_values(by = 'count', ascending = False).head(10)

```

### Visualization

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 18))

#Top 10 singers by number of songs
singer_df= singer_df.sort_values(by = 'count', ascending = False)
singer_df =singer_df.head(10)
sns.barplot(x ='singer' , y = 'count', data = singer_df,ax=axes[0,0] )
axes[0,0].set_title('Top 10 singers by number of songs', color = 'black')
axes[0,0].set_xlabel('Singers',color = 'black')
axes[0,0].set_ylabel('Number of songs', color = 'black')
axes[0, 0].tick_params(axis='x', rotation=90)


#Top 10 longest songs by duration
longest_songs = df.sort_values(by = 'duration' , ascending = False).head(10)
sns.barplot(x='name', y='duration', data= longest_songs,ax=axes[0,1])
axes[0,1].set_title('Top 10 Longest Songs by Duration')
axes[0,1].set_xlabel('Songs name')
axes[0,1].set_ylabel('Duration')
axes[0, 1].tick_params(axis='x', rotation=90)


#Punjabi songs by duration 3 to 4 mins
punjabi_songs = df[(df['language'] == 'Punjabi') & (df['duration']>=180) & (df['duration'] <=240)].head(10)
sns.barplot(x='name',y= 'duration', data=punjabi_songs,ax=axes[1,0])
axes[1,0].set_title('Punjabi Songs with Duration 3 to 4 Minutes')
axes[1,0].set_xlabel('Name')
axes[1,0].set_ylabel('Duration')
axes[1, 0].tick_params(axis='x', rotation=90)

#Top singer
sns.lineplot(data = singer_df , x= 'singer', y= 'count',ax=axes[1,1])
axes[1,1].set_title('Top singer')
axes[1,1].set_xlabel('Singer')
axes[1,1].set_ylabel('number of songs')
axes[1,1].tick_params(axis='x', rotation=90)


#Epic song of the year
epic_song= df.sort_values(by = 'duration' , ascending = False).head(5)
sns.barplot(data = epic_song , x= 'name', y= 'duration',ax=axes[2,0])
axes[2,0].set_title('Epic song of the year')
axes[2,0].set_xlabel('Songs')
axes[2,0].set_ylabel('Duration')
axes[2,0].tick_params(axis='x', rotation=90)


#Best collaboration
pair_counts = pair_counts.head()
sns.barplot(x='singer' , y= 'occurrences',data = pair_counts , ax=axes[2,1])
axes[2,1].set_title('Best Collaboration')
axes[2,1].set_xlabel('Singers collab')
axes[2,1].set_ylabel('Occurrences')
axes[2,1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
